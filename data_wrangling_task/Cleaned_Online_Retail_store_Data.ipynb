THE DATA ANALYSIS OF AN ONLINE RETAIL DATA.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



Online_retail=pd.read_csv(r"C:\Users\USER\Desktop\MENTORSHIP\WEEK TWO\Online Retail file.csv",encoding='unicode_escape')
Online_retail
```

We begin by first dropping duplicate data from the dataset

```python
Online_retail=Online_retail.drop_duplicates()
```

```python
Online_retail
```

the null rows are also filtered out with the dropna method

```python
Online_retail=Online_retail.dropna()
```

```python
Online_retail.dtypes
```

for statistical reasons the customer ID columns should not take any calculation hence we convert to string data type

```python
Online_retail['CustomerID']=Online_retail['CustomerID'].astype(str)
Online_retail['InvoiceDate']=Online_retail['InvoiceDate'].astype('datetime64[ns]')
```

```python
Online_retail['Quantity']=Online_retail['Quantity'].abs()
#the magnitude of quanity cannot be negative hence cleaning 
```

```python
Online_retail
```

```python
Online_retail['StockCode']=Online_retail['StockCode'].str.strip("QWERTYUIOPASDFGHJKLZXCVBNM")
# the stockcode need striping of number from letters as majority of them contains only number rows
```

```python
Online_retail
```

```python
Online_retail.dtypes
```

```python
Online_retail['InvoiceDate&Time']=Online_retail['InvoiceDate']
# not only the data is present in that column but also the time 
```

```python
Online_retail= Online_retail.drop('InvoiceDate', axis=1)
# we wpuld also confirm the data type with the 'dtype' method
```

```python
Online_retail.dtypes
```

```python
Online_retail
```

Since invoice some  rows in the invoice number column start with an alphabet it needs cleanings and also some rows in the stock no column cannot be blank hence replace with a zero

```python
Online_retail['InvoiceNo']=Online_retail['InvoiceNo'].str.strip("QWERTYUIOPASDFGHJKLZXCVBNM")
Online_retail['StockCode']=Online_retail['StockCode'].fillna(value=0)
```

The cleaned data is then cov=nverted to a csv format with the name Online_retail(updated)1.csv
