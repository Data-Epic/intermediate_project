# Handling inconsistent formatting
we first define a regex pattern, the match method is then used to filterout the incorrect format present in the the dataset


```python
# Define the patterns
invoice_no_pattern = r'^(C)?\d{6}$'
stock_code_pattern = r'^\d{5}$'
description_pattern = r'^(?!\d+$).+$'
quantity_pattern = r'^\d+$'
date_pattern = r'^\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}$'
unit_price_pattern = r'^\d+\.\d{2}$'
customer_id_pattern = r'^\d{5}$'
country_pattern = r'^[a-zA-Z\s]+$'
```

```python
# Check for inconsistencies in the InvoiceNo column
invoice_no = online_retail[~online_retail['InvoiceNo'].astype(str).str.match(invoice_no_pattern)]
invoice_no
```

```python
# Check for inconsistencies in the StockCode column
stock_code = online_retail[~online_retail['StockCode'].astype(str).str.match(stock_code_pattern)]
stock_code
```

```python
# Check for inconsistencies in the Description column
description = online_retail[~online_retail['Description'].astype(str).str.match(description_pattern)]
description
```

```python
# Check for inconsistencies in the Quantity column
quantity = online_retail[~online_retail['Quantity'].astype(str).str.match(quantity_pattern)]
quantity
```

```python
# Check for inconsistencies in the InvoiceDate column
date = online_retail[~online_retail['InvoiceDate'].astype(str).str.match(date_pattern)]
date
```

```python
# Check for inconsistencies in the UnitPrice column
unit_price = online_retail[~online_retail['UnitPrice'].astype(str).str.match(unit_price_pattern)]
unit_price
```

## HANDLING INCONSISTENCY

```python
new_online_retail = online_retail.copy()

# Remove rows with inconsistent InvoiceNo
new_online_retail = new_online_retail[new_online_retail['InvoiceNo'].astype(str).str.match(invoice_no_pattern)]

# Remove rows with inconsistent StockCode
new_online_retail['StockCode'] = new_online_retail['StockCode'].astype(str).str.extract('(\d{5})') #extract 5 digit numbers from StockCode
new_online_retail = new_online_retail.dropna(subset=['StockCode'])

# Convert UnitPrice to 2 decimal places
new_online_retail['UnitPrice'] = new_online_retail['UnitPrice'].round(2)

# Convert country names to title case
new_online_retail['Country'] = new_online_retail['Country'].str.title()

# Remove decimal place from CustomerID
new_online_retail['CustomerID'] = new_online_retail['CustomerID'].astype(str).str.split('.').str[0]

# Remove rows with inconsistent Description
new_online_retail = new_online_retail[new_online_retail['Description'].astype(str).str.match(description_pattern)] # remove rows with CustomerID containing only digits
new_online_retail['Description'] = new_online_retail['Description'].astype(str).str.strip().str.lower() # remove leading/trailing whitespaces and standardize to lowercase
```

```python
# reset index to delete the old index
new_online_retail.reset_index(drop=True, inplace=True)
new_online_retail.info()
```

```python
#Set original dataframe to new dataframe
online_retail = new_online_retail
online_retail.head()
```

Assingning correct datatype to the dataset.

```python
# set columns to correct data types

online_retail['InvoiceNo'] = online_retail['InvoiceNo'].astype(str)
online_retail['StockCode'] = online_retail['StockCode'].astype(str)
online_retail['Description'] = online_retail['Description'].astype(str)
online_retail['Quantity'] = online_retail['Quantity'].astype(int)
online_retail['InvoiceDate'] = pd.to_datetime(online_retail['InvoiceDate'])
online_retail['UnitPrice'] = online_retail['UnitPrice'].astype(float)
online_retail['CustomerID'] = online_retail['CustomerID'].astype(str)
online_retail['Country'] = online_retail['Country'].astype(str)
```

## Handling outliers in the UnitPrice and Quantity column
In the original datset some rows in the Quantity and Unitprice column have negatives, which ought not to be, we then use the abs() method to remove them and also set our row preference to positive integers

```python
online_retail['Quantity']=Online_retail['Quantity'].abs()
online_retail = Online_retail[Online_retail['Quantity'] > 0]
online_retail['UnitPrice']=online_retail['UnitPrice'].abs()
online_retail = online_retail[Online_retail['UnitPrice'] > 0]
online_retail.describe()
```

### Handling duplicates


```python
# drop duplicate rows present in the dataset
online_retail = online_retail.drop_duplicates()
online_retail.reset_index(drop=True, inplace=True)
```

Creating a new feature: Create a new Invoice data column reflecting the time too and also droping the old Invoice date column.
